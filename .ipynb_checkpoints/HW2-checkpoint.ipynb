Достаём из базы данных дневные и месячные сессии. Вдальнейшем будем работать только с месячными.

```python
df_daily, df_monthly = read_data("trade_info.sqlite3")
```

Группируем сессии по session_id и строим для них векторные представления путём дискретизации: нарезаем часовую сессию на фрагменты по минуте и из каждого берём последнюю стоимость. Для нормализации перед дискретизацией из каждой сессии вычтем средневзвешенное по ней. 

```python
groups = group_by_sesion_id(df_monthly)
vecs = vectorize_groups(groups)
```

Далее кластеризуем сессии с использованием разных метрик алгоритмом AgglomerativeClustering.
На графике разными цветами обозначены сессии, отнесенные к разным кластерам.

Сперва кластеризуем с помощью DTW. Отнормируем сессии по l2-норме. При увеличении количества кластеров можно пронаблюдать выявление новых паттернов, однако можно заметить невооруженным глазом, что в некоторых классах присутствуют выбросы.

```python
from tslearn.metrics import dtw

show_clusters_dynamics(vecs, groups, dist=dtw, norm=lambda x: x / np.linalg.norm(x, ord=2))
```

Затем кластеризуем данные с помощью косинусного расстояния между векторами, нормированными на l2-норму. Здесь паттерны кажутся более организованными, однако всё равно присутствуют выбросы.

```python
from sklearn.metrics.pairwise import cosine_distances as cosine

show_clusters_dynamics(vecs, groups,
    dist=lambda x, y: cosine([x], [y])[0][0],
    norm=lambda x: x / np.linalg.norm(x, ord=2))
```

Затем кластеризуем данные с помощью MAE и нормировкой на l1-норму. Здесь наблюдается что-то наоборот более хаотичное.

```python
from sklearn.metrics import mean_absolute_error as mae

show_clusters_dynamics(vecs, groups,
    dist=mae,
    norm=lambda x: x / np.linalg.norm(x, ord=1))
```

На мой взгляд при кластерном анализе месячных сессий, кластеризация с использованием косинусного расстояния выглядит наиболее осмысленно, хотя кластеризация с помощью DTW, вроде как единственная смогла распознать в сессии восходящий тренд в конце и отделить его от сессий с нисходящим.
